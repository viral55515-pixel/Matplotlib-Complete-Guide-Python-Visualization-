## SUBPLOTS IN MATPLOTLIB


```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

## OBJECT ORIENTED WAY

#### matplotlib allows the aspect ratio dpi and figure size to the specifed when the figgure object is created. you can use the figsize and dpi keyword arguments.

#### fig size is a tuple of the width and height of the figure in inches dpi is the dots per inch PPI(points per inch)

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis1 = fig.add_axes([0.3,0.4,0.3,0.3])
axis1.plot(x,y)
```

```python
fig = plt.figure(figsize=(20,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis1 = fig.add_axes([0.3,0.4,0.3,0.3])
axis1.plot(x,y)
```

## BOXPLOT AND VIOLINPLOT
### BOXPLOT AND VIOLIONPOTS ARE USED TOSHOWN THE DISTRIBUTION OF CATEGORICAL DATA. A BOX PLOT (OR BOX- AND - WHISKER PLOT)SHOWS THE DISTRIBUTIN OF QUANTITATIVE DATA IN A WAY THAT FACILATE COMPARISONS BETWEEN VARIBLES OR ACROSS LEVELS OF A CATEGORICAL VARIABLE.THE BOX SOWS THE QUARTILES OF THE DATASET WHILE THE WHISKERS EXTEND TO SHOW THE REST OF THE DISTRIBUTION,EXPECT FOR POINTS USING A METHOD THAT IS A FUNCTION OF THE INTER QUARTILE RNGE.

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

## STRIPPLOT SWARMPLOT
### The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

### The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
import seaborn as sns



# Create the plot
sns.stripplot(x='day', y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x='tip', y='day', data=df)
sns.stripplot(x='tip', y='day', data=df)
```
